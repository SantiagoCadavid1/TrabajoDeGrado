```python
p=-1
signalp_list=[]
signaln_list=[]
for i in range(len(r_list)):
    if((i)%5==0):
        p=p+1
    p_dir="data/signals/p"+chr((p_list[p]//10)+48)+chr((p_list[p]%10)+48)+"r1.mat"
    x=loadmat(p_dir)
    ecg=(x['val']-0)/200
    ecg=np.transpose(ecg)
    fs=500
    ts=1/fs
    r=r_list[i]
    nr=nr_list[i]
    signalp_list.append(ecg[r-25:r+25])
    signaln_list.append(ecg[nr-25:nr+25])
```

```python
import json

json.dumps(np.transpose(signalp_list[0])[0].tolist())
```

```python
import csv

datos=[]
for i in signalp_list:
    datos.append([json.dumps(np.transpose(i)[0].tolist()),1])
    
for i in signaln_list:
    datos.append([json.dumps(np.transpose(i)[0].tolist()),0])
    
with open('data/datos.csv', 'w', newline='') as archivo:
    writer = csv.writer(archivo)
    writer.writerows(datos)
```

```python
with open('data/datos.csv', newline='') as archivo:
    reader = csv.reader(archivo)

    # Leemos los datos del archivo CSV y los convertimos a su tipo original
    datos = []
    for fila in reader:
        # Convertimos la primera columna de la fila en un arreglo num√©rico
        arreglo_leido = json.loads(fila[0])
        etiqueta = json.loads(fila[1])
        # Agregamos la fila convertida a la lista de datos
        datos.append([np.transpose(arreglo_leido), etiqueta])
```

```python
import pandas as pd

datos = pd.read_csv('data/datos.csv')
```

```python
signal_list=[]
for i in range(len(datos)):
    signal_list.append(json.loads(datos.iloc[i, 0]))
```

```python
db_mean=np.mean(signal_list)
db_std=np.std(signal_list)
db_min=np.min(signal_list)
db_max=np.max(signal_list)
```

```python
norm_vars = {'mean': db_mean, 'std': db_std, 'min': db_min, 'max': db_max}

# Guardar el diccionario en un archivo JSON
with open('data/variables.json', 'w') as f:
    json.dump(norm_vars, f)
```
